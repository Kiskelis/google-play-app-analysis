```python
# Load datasets from HuggingFace
mbr_data = load_dataset('recmeapp/mobilerec', data_dir='interactions')
mbr_meta = load_dataset('recmeapp/mobilerec', data_dir='app_meta')

# Convert both to pandas
interactions_df = mbr_data['train'].to_pandas()
meta_df = mbr_meta['train'].to_pandas()

# Save separate datasets
interactions_df.to_csv('mobilerec_interactions.csv', index=False)
meta_df.to_csv('mobilerec_metadata.csv', index=False)

# Merge on app_package
full_df = pd.merge(interactions_df, meta_df, on='app_package', how='left')

# Save merged dataset
full_df.to_csv('mobilerec_full_merged.csv', index=False)
```
